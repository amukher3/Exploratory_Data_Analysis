```python
temp=df[ 'acousticness'].mean()
print(f'Average accousticness is:{temp}')
temp=df['danceability'].mean()
print(f'Average danceability is:{temp}')
temp=df['duration_ms'].mean()
print(f'Average duration is:{temp}')
temp=df['energy'].mean()
print(f'Average energy is:{temp}')
temp=df['loudness'].mean()
print(f'Average Loudnes is:{temp}')
temp=df['popularity'].mean()
print(f'Average popularity is:{temp}')
```

```python
def topK(k,dictionary):
    
    artistDict=dict()
    topkDict=dict()
    tempPopularity=0
    
    for i in dictionary:
        for j in range(len(df['artists'])):
            if df['artists'][j]==i:
                tempPopularity=df['popularity'][j]
                
                if df['artists'][j] not in artistDict.keys():
                    artistDict.update({i:tempPopularity})
                else:
                    artistDict[i]+=tempPopularity
                    
    #the top k artists by popularity: 
    tempTup=sorted(artistDict.items() , key=lambda x: x[1],reverse=True)
    for i in range(k):
        topkDict.update({tempTup[i][0]:tempTup[i][1]})
        
    return topkDict
        
```

```python
#For classical genres:

classicalDict={'Average Danceability':0,'Average duration':0,'Average energy':0,
               'Average tempo':0,'Average Valence':0,
               'Average popularity':0,'Average loudness':0,'artist':[]}

for i in range(len(df['genres'])):
    if 'classical' in df['genres'][i] or 'Classical' in df['genres'][i] :
        classicalDict['Average Danceability']+=df['danceability'][i]/len(df['genres'])
        classicalDict['Average duration']+=df['duration_ms'][i]/len(df['genres'])
        classicalDict['Average energy']+=df['energy'][i]/len(df['genres'])
        classicalDict['Average tempo']+=df['tempo'][i]/len(df['genres'])
        classicalDict['Average Valence']+=df['valence'][i]/len(df['genres'])
        classicalDict['Average popularity']+=df['popularity'][i]/len(df['genres'])
        classicalDict['Average loudness']+=df['loudness'][i]/len(df['genres'])
        classicalDict['artist'].append(df['artists'][i])
```

```python
#For classical genres
#top-10 artists based on popularity score

tempDict=topK(10,classicalDict['artist'])

for key,vals in tempDict.items():
    print(f'Artist {key} with a total popularity score of {vals}')
```
